```python
with open("Kurs.ipynb") as f:
    s = f.read()
```

```python
# Ny blokk så vi slipper å lese alt på nytt hele tiden.

len(s)

# 1549 bokstaver. Formatert som JSON.
# help(json)

json.loads(s)

# ser bra ut!

cells = json.loads(s)["cells"] # Vi må kunne dictionaries!

len(cells)

for cell in cells:
    print(cell.keys())
```

```python
for cell in cells:
    print(cell["cell_type"])
    if cell["cell_type"] == "code":
        print(cell["source"])
```
